```python

#la corrlation n'est pas super, 
#on va regarder toutes les corrlations possibles pour trouver la meilleure

df3=df.copy()
L=['Life Ladder','Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption','Positive affect','Negative affect']

def max(n,m):
    if n<m:
        return m
    return n

maxi=0
V1,V2='',''
for variable1 in L:
    for variable2 in L:
        if variable1 !=variable2:
            df4=df3[['Country name', variable1, variable2]]
            pays_moyenne = df4.groupby(by='Country name').mean()
            #print(pays_moyenne[variable1])
            correlation_coefficient = pays_moyenne[variable1].corr(pays_moyenne[variable2])
            #print(correlation_coefficient)
            maxi=max(maxi,correlation_coefficient)
            if maxi==correlation_coefficient:
                V1=variable1
                V2=variable2
print(maxi,V1,V2)
#Rsultat: 0.8349472534887966, Log GDP per capita, Life Ladder
#c'est logiquement le pib par habitant qui influence life ladder
#pas trs intrssant, les deux sont des notions d'argent

#essayons sans life ladder

```
